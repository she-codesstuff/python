### SETS
- undered so no indexes and insertion order
- has no duplicate elements
- is based on data structure hash table

```python
## Defining an empty set

set_var = set()
print(set_var)
print(type(set_var))
```

```python
set_var =  {"avengers","ironman","hitman"}
print(set_var)
type(set_var)
```

```python
# Indexing is not supportable
set_var[0]
```

```python
set_var["avengers"]
```

```python
# Inbuilt function in sets
set_var.add("hulk")
print(set_var)
```

```python
set1 = {'avengers', 'hulk', 'ironman', 'hitman'}
set2 = {'avengers', 'hulk', 'ironman', 'hitman','hulk2'}
```

```python
set2.intersection(set1) # ;ooks for common elements/items
```

```python
#difference
set2.difference(set1)
```

```python
# updation
set2.difference_update(set1)
print(set2)
```

## Dictionaries
- key and values
- unordered
- indexed

```python
dic={}
```

```python
type(dic)
```

```python
set_ex = {1,2,3,4,5}
type(set_ex) # we have to add values
```

```python
# adding manually
set_ex = {1:"apple",2:"ball",3:"cat"}
type(set_ex)
```

```python
## Accessing the item
set_ex[1]
```

```python
## using loops (keys)
for x in set_ex:
    print(x)
```

```python
## using loops (values)
for x in set_ex.values():
    print(x)
```

```python
## for both 
for x in set_ex.items():
    print(x)
```

```python
## adding items in dictionaries
set_ex[4] = 'dog'
set_ex
```

```python
set_ex[1] = 'aerospace' # replacing
set_ex
```

#### Nested Dictionary

```python
car1_model = {1:"apple"}
car2_model    = {  2:"ball"}
car3_model   ={   3:"cat"}
car_type = {'ex1':car1_model,'ex2':car2_model,'ex3':car3_model}
print(car_type)
```

```python
print(car_type['ex1'])
```

```python
print(car_type['ex1'][1])
```
