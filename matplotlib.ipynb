### MatplotLib

```python
import numpy as np
import matplotlib.pyplot as plt  ### here no need to write plt on each time
%matplotlib inline
```

```python
## Simple example
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
### plotting using matplotlib
## plt scater
plt.scatter(x,y,c='r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("graphs in 2D")
plt.fig('test.png')## had to write plt.show()
```

```python
y=x*x
```

```python
###plotting

plt.plot(x,y,'b--')
```

```python
# plt.plot(x,y,'b+-')
# plt.plot(x,y,'b*-')
plt.plot(x,y,'b+',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('2d diagram')
```

```python
### creating subplots
plt.subplot(2,2,1)# rows columns and position
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,3)
plt.plot(x,y,'g')
plt.subplot(2,2,4)
plt.plot(x,y,'y--')
```

```python
x = np.arange(1,11)
y = 3*x+5
plt.plot(x,y,'b*-',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('matplotlib demo')
```

```python
# Compute x and y 
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave")
plt.plot(x,y)
```

```python
#Subplot
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 7 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
# Bar plot
x=[2,3,10]
y = [11,16,1]
x2=[1,2,5,6]
y2=[8,9,7,8]
plt.bar(x,y)
plt.bar(x2,y2,color = 'r')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('matplotlib demo')
```

### Histograms

```python
a = np.array([22,3,4,5,5,6,3,45,6,34,56,7,8,90,34,5,6])
plt.hist(a)
plt.title("histogram") #//it gives the density
```

### Box Plot using Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True) #it starts from 0,25,50,75,100 
```

### Pie chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0.2, 0.1, 0.2)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)
# for floting formats
plt.axis('equal')
plt.show()
```
